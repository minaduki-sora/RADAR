```python
from datasets import load_from_disk
import numpy as np
from sklearn.feature_selection import mutual_info_classif, f_classif
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 载入数据
dataset = load_from_disk("../data/mt-bench-llama3-d13-topk10-t0-cal/arrow")["train"]
dataset.set_format("numpy")
X = np.array(dataset["features"])
y = np.array(dataset["accept_length"])

# 互信息
mi = mutual_info_classif(X, y, discrete_features=False, random_state=42)
print("互信息:")
for idx, score in enumerate(mi):
    print(f"特征{idx} 与标签的互信息: {score:.4f}")

# 方差分析
f_scores, p_values = f_classif(X, y)
print("\nF值与p值:")
for idx, (f, p) in enumerate(zip(f_scores, p_values)):
    print(f"特征{idx} F值: {f:.4f}, p值: {p:.4e}")

# 随机森林重要性
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X, y)
importances = clf.feature_importances_
print("\n随机森林特征重要性:")
for idx, imp in enumerate(importances):
    print(f"特征{idx} 重要性: {imp:.4f}")

# 可视化（可选，特征数多时建议只画top3）
df = pd.DataFrame(X, columns=[f"f{i}" for i in range(X.shape[1])])
df["label"] = y
top_k = np.argsort(mi)[-3:]
for i in top_k:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x="label", y=f"f{i}", data=df)
    plt.title(f"特征{i}在不同标签下的箱线图")
    plt.show()
```
