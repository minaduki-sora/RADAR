```python
from datasets import load_from_disk
import numpy as np
from sklearn.feature_selection import mutual_info_classif, f_classif
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 载入数据
dataset = load_from_disk("../data/mt-bench-llama3-d9-topk10-t0-cal/arrow")["train"]
dataset.set_format("numpy")
X = np.array(dataset["features"])
y = np.array(dataset["accept_length"])

# 互信息
mi = mutual_info_classif(X, y, discrete_features=False, random_state=42)
print("互信息:")
for idx, score in enumerate(mi):
    print(f"特征{idx} 与标签的互信息: {score:.4f}")

# 方差分析
f_scores, p_values = f_classif(X, y)
print("\nF值与p值:")
for idx, (f, p) in enumerate(zip(f_scores, p_values)):
    print(f"特征{idx} F值: {f:.4f}, p值: {p:.4e}")

# 随机森林重要性
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X, y)
importances = clf.feature_importances_
print("\n随机森林特征重要性:")
for idx, imp in enumerate(importances):
    print(f"特征{idx} 重要性: {imp:.4f}")

# 可视化（可选，特征数多时建议只画top3）
df = pd.DataFrame(X, columns=[f"f{i}" for i in range(X.shape[1])])
df["label"] = y
top_k = np.argsort(mi)[-3:]
for i in top_k:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x="label", y=f"f{i}", data=df)
    plt.title(f"特征{i}在不同标签下的箱线图")
    plt.show()
```

```python
# analyze scores
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from datasets import load_from_disk
```

```python
# 读取数据集
dataset = load_from_disk("../data/scores/mt-bench-llama3-d9-topk10-t0")["train"]

# 准备收集所有层的数据
eagle_scores = {f"eagle_{i}_forward": [] for i in range(1, 11)}

# 遍历数据集，将每层的置信分数合并到一个大list中
for sample in dataset:
    for i in range(1, 11):
        key = f"eagle_{i}_forward"
        eagle_scores[key].extend(sample[key])

# 转为numpy数组加速后续处理
for k in eagle_scores:
    eagle_scores[k] = np.array(eagle_scores[k], dtype=np.float16)

# 可视化：每层各自的密度分布
plt.figure(figsize=(20, 10))
for i in range(1, 11):
    key = f"eagle_{i}_forward"
    sns.kdeplot(eagle_scores[key], label=key)
plt.title("Eagle Forward Token Score Distributions (各层置信分数分布)")
plt.xlabel("Score")
plt.ylabel("Density")
plt.legend()
plt.show()

# 统计描述
stat_df = pd.DataFrame({
    key: [
        eagle_scores[key].mean(),
        eagle_scores[key].std(),
        np.percentile(eagle_scores[key], 25),
        np.percentile(eagle_scores[key], 50),
        np.percentile(eagle_scores[key], 75)
    ]
    for key in eagle_scores
}, index=['mean', 'std', 'q25', 'median', 'q75']).T

print(stat_df)

# 箱型图展示
plt.figure(figsize=(12, 6))
sns.boxplot(data=[eagle_scores[f'eagle_{i}_forward'] for i in range(1, 11)])
plt.xticks(ticks=range(10), labels=[f'eagle_{i}_forward' for i in range(1, 11)], rotation=45)
plt.ylabel('Score')
plt.title('Boxplot of Eagle Forward Token Scores (Each Layer)')
plt.show()
```

## 分析不同层max scores的分布

```python
# 分析argmax
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_from_disk

# 加载数据集
dataset = load_from_disk("../data/scores/mt-bench-llama3-d9-topk10-t0")["train"]

# 构建数据框
records = []
for sample in dataset:
    accept_length = sample["accept_length"]
    for n in range(1, 11):
        layer = n
        max_score = np.max(sample[f"eagle_{n}_forward"])
        records.append({
            "layer": layer,
            "max_score": max_score,
            "accept_length": accept_length
        })

df = pd.DataFrame(records)

# 画图, 每一层分别画，按accept_length分组
fig, axes = plt.subplots(2, 5, figsize=(25, 8), sharey=True)
for i in range(1, 11):
    ax = axes[(i-1)//5, (i-1)%5]
    sns.violinplot(
        data=df[df["layer"] == i], 
        x="accept_length", 
        y="max_score", 
        ax=ax
    )
    ax.set_title(f"Layer {i}")
    ax.set_xlabel("accept_length")
    ax.set_ylabel("max_score")
plt.suptitle("Max Score Distribution per Layer by Accept Length", fontsize=20)
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
