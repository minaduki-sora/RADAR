```python
import json
from collections import Counter
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.ticker import PercentFormatter

# 设置PDF保存参数
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

# 设置Times Roman字体
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman', 'Times', 'DejaVu Serif']
plt.rcParams['mathtext.fontset'] = 'stix'

def load_action_lengths(file_path, is_eagle=False):
    """从JSONL文件中加载action_lengths数据"""
    all_action_lengths = []
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            try:
                data = json.loads(line)
                for choice in data.get("choices", []):
                    # 处理EAGLE-3的特殊结构
                    if is_eagle:
                        if "accept lengths" in choice:
                            # 展平二维列表
                            for sublist in choice["accept lengths"]:
                                all_action_lengths.extend(sublist)
                    # 处理RADAR的标准结构
                    else:
                        if "action_lengths" in choice:
                            all_action_lengths.extend(choice["action_lengths"])
            except json.JSONDecodeError:
                continue
    return all_action_lengths

def calculate_frequencies(action_lengths):
    """计算action lengths的频率分布"""
    if not action_lengths:
        print("警告：未找到任何action_lengths数据")
        return None, None
    
    frequency_counter = Counter(action_lengths)
    min_val = min(frequency_counter.keys())
    max_val = max(frequency_counter.keys())
    numbers = list(range(min_val, max_val + 1))
    sum_counts = sum(frequency_counter.values())
    frequencies = [frequency_counter.get(num, 0)/sum_counts for num in numbers]
    return numbers, frequencies

# 文件路径 - 请替换为实际路径
eagle_file = '/home/majunjie/code/EAGLE/output/mt_bench/llama3.1/t1d7/eagen-3-acclen.jsonl'
radar_file = '/home/majunjie/code/EAGLE/output/mt_bench/llama3.1/t1d7/time/hawkeye-b-0.03084-a-0.02000-g-0.99-lr-1e-04-wd-1e-04-dr-0.10-actions-3.jsonl'  # 替换为实际的RADAR文件路径

# 加载数据（指定is_eagle参数）
eagle_lengths = load_action_lengths(eagle_file, is_eagle=True)
radar_lengths = load_action_lengths(radar_file, is_eagle=False)

# 计算频率分布
eagle_numbers, eagle_freqs = calculate_frequencies(eagle_lengths)
radar_numbers, radar_freqs = calculate_frequencies(radar_lengths)

# 确保有有效数据
if eagle_freqs is None or radar_freqs is None:
    exit()

# 确定共同的范围
min_val = min(min(eagle_numbers), min(radar_numbers))
max_val = max(max(eagle_numbers), max(radar_numbers))
all_numbers = list(range(min_val, max_val + 1))

# 创建图表
plt.figure(figsize=(14, 8))

# 绘制EAGLE-3数据
eagle_bars = plt.bar(
    [x - 0.2 for x in all_numbers],
    [eagle_freqs[all_numbers.index(x)] if x in eagle_numbers else 0 for x in all_numbers],
    width=0.4, 
    color="#ebb625", 
    edgecolor='black', 
    alpha=0.8,
    label='EAGLE-3'
)

# 绘制RADAR数据
radar_bars = plt.bar(
    [x + 0.2 for x in all_numbers],
    [radar_freqs[all_numbers.index(x)] if x in radar_numbers else 0 for x in all_numbers],
    width=0.4, 
    color="#1f77b4",  # 蓝色
    edgecolor='black', 
    alpha=0.8,
    label='RADAR'
)

# 图表美化
# plt.xlabel('Acceptance Lengths', fontsize=30, labelpad=26, fontweight='bold')
plt.ylabel('Frequency', fontsize=30, labelpad=26, fontweight='bold')

# 设置y轴为百分比格式
plt.gca().yaxis.set_major_formatter(PercentFormatter(1.0))

# 设置刻度标签大小
plt.xticks(all_numbers, fontsize=28)
plt.yticks(fontsize=28)

# 添加图例
plt.legend(fontsize=24, loc='upper right')

# 设置网格线
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 调整布局
plt.tight_layout(pad=3.0)

# 保存为PDF矢量图
pdf_path = '/home/majunjie/code/EAGLE/output/acccomp.pdf'
plt.savefig(pdf_path, format='pdf', bbox_inches='tight')
print(f"\n矢量图已保存至: {pdf_path}")

# 可选：关闭图形显示
plt.close()
```

```python
import json
from collections import Counter
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.ticker import PercentFormatter
import numpy as np

# 设置PDF保存参数（矢量图）
matplotlib.rcParams['pdf.fonttype'] = 42  # 确保文本可编辑
matplotlib.rcParams['ps.fonttype'] = 42   # 防止Type 3字体

# 设置Times Roman字体
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman', 'Times', 'DejaVu Serif']
plt.rcParams['mathtext.fontset'] = 'stix'  # 数学公式字体

# 初始化计数器
all_action_lengths = []

# 读取JSONL文件
jsonl_file = '/home/majunjie/code/EAGLE/output/mt_bench/llama3.1/t1d7/time/hawkeye-b-0.03084-a-0.02000-g-0.99-lr-1e-04-wd-1e-04-dr-0.10-action_d-3.jsonl'

with open(jsonl_file, 'r', encoding='utf-8') as f:
    for line in f:
        try:
            data = json.loads(line)
            for choice in data.get("choices", []):
                if "action_lengths" in choice:
                    all_action_lengths.extend(choice["action_lengths"])
        except json.JSONDecodeError:
            continue  # 静默跳过错误行

# 统计频率
if not all_action_lengths:
    print("未找到任何action_lengths数据")
    exit()

frequency_counter = Counter(all_action_lengths)

# 准备绘图数据
min_val = min(frequency_counter.keys())
max_val = max(frequency_counter.keys())
numbers = list(range(min_val, max_val + 1))
sum_counts = sum(frequency_counter.values())
frequencies = [frequency_counter.get(num, 0)/sum_counts for num in numbers]

# 添加EAGLE-3数据（固定调用次数为8）
eagle3_value = 8
if eagle3_value not in numbers:
    numbers.append(eagle3_value)
    frequencies.append(0)  # 当前方法在8处的频率为0
    # 重新排序
    numbers, frequencies = zip(*sorted(zip(numbers, frequencies)))
    numbers = list(numbers)
    frequencies = list(frequencies)

# 创建EAGLE-3的频率数据（100%在8的位置）
eagle3_freq = [0] * len(numbers)
if eagle3_value in numbers:
    eagle3_freq[numbers.index(eagle3_value)] = 1.0

# 创建图表
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), 
                               gridspec_kw={'height_ratios': [1, 3]},
                               sharex=True)
plt.subplots_adjust(hspace=0.05)  # 减少子图间距

# 设置x轴位置
x = np.arange(len(numbers))
bar_width = 0.35

# 在下部主图中绘制当前方法的数据（0-20%范围）
ax1.bar(x - bar_width/2, frequencies, width=bar_width, color='#1f77b4', edgecolor='black', alpha=0.8, label='Current Method')
ax1.set_ylim(0, 0.2)  # 只显示0-20%的范围
ax1.set_ylabel('Frequency', fontsize=30, labelpad=26, fontweight='bold')
ax1.yaxis.set_major_formatter(PercentFormatter(1.0))
ax1.tick_params(axis='y', labelsize=28)
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# 在上部小图中绘制EAGLE-3的数据（80-100%范围）
ax2.bar(x + bar_width/2, eagle3_freq, width=bar_width, color='#ebb625', edgecolor='black', alpha=0.8, label='EAGLE-3')
ax2.set_ylim(0.8, 1.0)  # 只显示80-100%的范围
ax2.tick_params(axis='y', labelsize=28)
ax2.grid(axis='y', linestyle='--', alpha=0.7)

# 隐藏下部图的顶部边框和上部图的底部边框
ax1.spines['top'].set_visible(False)
ax2.spines['bottom'].set_visible(False)
ax1.xaxis.tick_bottom()
ax2.xaxis.tick_top()
ax2.tick_params(labeltop=False)  # 不在上部图显示x轴标签

# 添加折断符号
d = 0.015  # 折断符号的大小
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)        # 左下折断符号
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # 右下折断符号

kwargs.update(transform=ax2.transAxes)  
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # 左上折断符号
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # 右上折断符号

# 设置x轴标签和刻度
plt.xlabel('Number of Calls', fontsize=30, labelpad=26, fontweight='bold')
plt.xticks(x, numbers, fontsize=28)

# 添加图例
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
fig.legend(handles1 + handles2, labels1 + labels2, 
           loc='upper right', fontsize=24, framealpha=1.0)

# 调整布局
plt.tight_layout(pad=3.0)
plt.subplots_adjust(top=0.9)  # 为图例留出空间

# 保存为PDF矢量图
pdf_path = '/home/majunjie/code/EAGLE/output/action_lengths_frequency.pdf'
plt.savefig(pdf_path, format='pdf', bbox_inches='tight')
print(f"\n矢量图已保存至: {pdf_path}")

# 可选：关闭图形显示
# plt.close()
```
